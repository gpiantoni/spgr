## Wavelet Transform

### Theoretical Ideas

Recordings on the grid provide information with high spatial resolution.

To obtain information of spindles in the time and frequency domains, we transformed the data using the wavelet transformation. 
The wavelets are clearly localized in time and frequency.

```python
from phypno.trans.frequency import morlet
from matplotlib.pyplot import plot, xlabel, ylabel, xlim, ylim
from numpy import real, imag, arange

lg = getLogger('phypno')
lg.setLevel(DEBUG)
```

The wavelet had center frequency at 13 Hz (but maybe we need to adapt it to each electrode / patient after detecting the peak in spectrogram).
The standard deviation in the frequency domain was 2 Hz, so that 95% of the power is concentrated between 9 and 17 Hz.
The standard deviation in the time domain was conversevely 80 ms, so that 95% of the energy was in the window 320ms.
Wavelets were normalized to have energy equal to 1.

```python
frequency = 13
sigma_f = 2
s_freq = 1024
```

```python
w = morlet(frequency, s_freq, sigma_f=sigma_f, normalization='area')

t = arange(0, w.shape[0], 1 / s_freq)
plot(real(w))
plot(imag(w), 'r')
xlabel('time (s)')
ylabel('wavelet coef (A.U.)')
```

Wavelets are localized in the spindle band in the frequency domain

```python
from scipy.signal import periodogram

[f, Pxx] = periodogram(w, s_freq, nfft=s_freq)
plot(f, Pxx)
xlim(0, 20)
xlabel('frequency (Hz)')
ylabel('wavelet coef (A.U.)')
```

### Application to one epoch

```python
import matplotlib
matplotlib.matplotlib_fname()
import matplotlib as mpl
mpl.rcParams['figure.figsize'] = (18, 4)
```

```python
from spgr.read_data import list_subj, get_data, GROUP_DIR
STAGE = 'sleep'
from phypno.trans import Select, TimeFreq
```

```python
subj = 'EM09'
TRIAL = 25
chan = 'GR29'

full_data = get_data(subj, 'sleep', ('grid', ), reref='avg', resample_freq=None)
```

Recordings were scored on 30s epochs and the analysis was conducted on each epoch.
Consecutive epochs were not concatenated (but maybe they should be).

```python
sel = Select(trial=(TRIAL, ), time=(full_data.axis['time'][TRIAL][0], full_data.axis['time'][TRIAL][-2]))
data = sel(full_data)
```

```python
x = data.axis['time'][0]
y = data(trial=0, chan=chan)
plot(x, y)
ylim(-200, 200)

xlabel('Time')
ylabel('Amplitude (uV)')
```

```python
plot(x, y)
xlim(x[0] + 20, x[0] + 30)
ylim(-200, 200)

xlabel('Time')
ylabel('Amplitude (uV)')
```

The wavelet transform transforms the signal into its real and imaginary part.

```python
calc_wavelet = TimeFreq(foi=(frequency, ), sigma_f=sigma_f, normalization='area')
tf = calc_wavelet(data)
```

```python
y = tf(trial=0, chan=chan)
plot(x, real(y))
plot(x, imag(y))
xlim(x[0] + 20, x[0] + 30)
ylim(-200, 200)

xlabel('Time')
ylabel('Amplitude (A.U.)')
```

```python
from numpy import angle, abs, pi, diff
```

From the complex signal, we can obtain the absolute value (which is the instantenuous power) and the instantenous phase.

```python
plot(x, abs(y))
xlim(x[0] + 20, x[0] + 30)
ylim(0, 250)

xlabel('Time')
ylabel('Amplitude')
```

```python
plot(x, angle(y))
xlim(x[0] + 20, x[0] + 30)
ylim(-pi, pi)

xlabel('Time')
ylabel('Amplitude')
```

By taking the first derivative of the instaneous phase (which for a discrete signal is equal to taking the difference between consecutive points), we obtain the instaneous frequency.
Discontinuities in the instaneous frequency indicates a phase reset.

```python
inst_freq = diff(angle(y), axis=0) / pi * data.s_freq / 2
inst_freq[inst_freq < 0] += data.s_freq  # rectify in the opposite direction

plot(x[:-1], inst_freq)
xlim(x[0] + 20, x[0] + 30)
ylim(0, 30)

xlabel('Time')
ylabel('Phase')
```

It appears that the choice of the wavelet parameters can influence the position and charateristics of the discontinuities.
